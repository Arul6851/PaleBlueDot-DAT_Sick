# AOI
#### The Agriculture Orientation Index (AOI) for Government Expenditures is defined as the Agriculture Share of Government Expenditures, divided by the Agriculture Share of GDP, where Agriculture refers to the agriculture, forestry, fishing and hunting sector.      
#### The measure is a currency-free index, calculated as the ratio of these two shares.       
#### This indicator will measure progress towards SDG Target 2.a.

## More clear explanation:
The Agriculture Orientation Index (AOI) is a way to measure how much a government focuses its financial resources on agriculture compared to the overall economic output of the country. It is calculated by dividing the percentage of government spending dedicated to agriculture by the percentage of the Gross Domestic Product (GDP) generated by agriculture. The higher the AOI, the more the government is prioritizing and investing in agriculture relative to its contribution to the overall economy.

## <u> Start of EDA

```python
import pandas as pd
```

```python
df_AOI = pd.read_csv('/kaggle/input/zerohunger/agriculture-orientation-index.csv')
```

```python
df_AOI.head(20)
```

```python
df_AOI = df_AOI.drop('Code', axis=1)
```

```python
df_AOI.rename(columns={'Entity': 'Country'}, inplace=True)
df_AOI.rename(columns={'Credit to Agriculture, Forestry and Fishing | 00023068 || Agriculture orientation index US$ | 006159 || Index': 'Index'}, inplace=True)
```

```python
df_AOI.head(20)
```

```python
df_AOI.isnull().sum()
# Therefore no null values
```

```python
df_AOI['Country'].unique()
```

```python
countries = df_AOI['Country'].unique()

# Filter countries with (FAO) in their names
fao_list = [country for country in countries if '(FAO)' in country]
fao_list
```

Removing the Countries that has (FAO) and 'world' so that it is easier to vizualize.

```python
# Add 'World' to the list of countries to be dropped
countries_to_drop = fao_list + ['World']

# Filter out rows where 'Country' is in countries_to_drop
df_AOI_filtered = df_AOI[~df_AOI['Country'].isin(countries_to_drop)]


df_AOI_filtered

# df_AOI_filtered has the data that has no world or countries with FAO
```

```python
# Filter countries with at least one 'Index' value less than 1 in 2020 or 2021
selected_countries = df_AOI_filtered[(df_AOI_filtered['Year'] == 2020) | (df_AOI['Year'] == 2021)]['Country'][df_AOI['Index'] < 1].unique()

# Create df_AOI_lessThan1 by filtering the data for the selected countries
df_AOI_lessThan1 = df_AOI_filtered[df_AOI_filtered['Country'].isin(selected_countries)]

df_AOI_lessThan1
```

```python
(df_AOI_lessThan1['Index'] > 1).any()
```

```python
df_AOI_lessThan1[df_AOI_lessThan1['Index'] > 1]['Country'].unique()

## This bascically shows the countries that had AOI more that 1 at one point but right now the AOI is less than 1
```

```python
# Group by 'Country' and calculate the average decrease
avg_decrease = df_AOI_lessThan1.groupby('Country')['Index'].apply(lambda x: x.iloc[-1] - x.iloc[0]).reset_index(name='AverageDecrease')

# Sort the countries based on the average decrease
sorted_countries = avg_decrease.sort_values(by='AverageDecrease', ascending=False)

# Select the top 10 countries
top_10_countries = sorted_countries.head(10)

# Display the result
print(top_10_countries[['Country', 'AverageDecrease']])
```
