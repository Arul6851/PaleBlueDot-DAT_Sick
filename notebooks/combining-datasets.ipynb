### So it is better if we use values from 2001 to 2020
So drop all values before 2001 and after 2020        
Also fill nan values for empty values within 2001 to 2020

```python
for df_name, df in dataframes.items():
    # Check if 'Year' column exists in the DataFrame
    if 'Year' in df.columns:
        # Filter data between 2001 and 2020
        dataframes[df_name] = df[(df['Year'] >= 2001) & (df['Year'] <= 2020)]

        # Update the original DataFrame outside the loop
        globals()[df_name] = dataframes[df_name]
```
