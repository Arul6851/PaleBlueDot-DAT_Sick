```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_aoi = pd.read_csv('/kaggle/input/zerohunger/agriculture-orientation-index.csv')
df_water = pd.read_csv('/kaggle/input/pale-blue-dot/implementation-of-integrated-water-resource-management.csv')
df_lit = pd.read_csv('/kaggle/input/pale-blue-dot/cross-country-literacy-rates.csv')
df_mal = pd.read_csv('/kaggle/input/pale-blue-dot/prevalence-of-undernourishment.csv')
```

```python
df_water.rename(columns={'Entity': 'Country'}, inplace=True)
df_water.rename(columns={'6.5.1 - Degree of integrated water resources management implementation (%) - ER_H2O_IWRMD':'water'}, inplace=True)

df_aoi.rename(columns={'Entity': 'Country'}, inplace=True)
df_aoi.rename(columns={'Credit to Agriculture, Forestry and Fishing | 00023068 || Agriculture orientation index US$ | 006159 || Index':'aoi'}, inplace=True)

df_mal.rename(columns={'Entity': 'Country'}, inplace=True)
df_mal.rename(columns={'Prevalence of undernourishment (% of population)':'undernourishment'}, inplace=True)

df_lit.rename(columns={'Entity': 'Country'}, inplace=True)
df_lit.rename(columns={'Historical and more recent literacy estimates':'literacy'}, inplace=True)
```

```python
merged_df = pd.merge(df_aoi, df_mal, on=['Country','Code', 'Year'], how='outer')
merged_df = pd.merge(merged_df, df_water, on=['Country','Code', 'Year'], how='outer')
merged_df = pd.merge(merged_df, df_lit, on=['Country','Code', 'Year'], how='outer')
```

```python
merged_df.shape
```

```python
merged_df
```

```python
merged_df = merged_df[merged_df['Year']==2020]
# merged_df.iloc[100]
```

```python
sorted_merged_df = merged_df.sort_values(by='aoi', ascending=True)
sorted_merged_df
```
